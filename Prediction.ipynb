```python
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import load_model
import pickle
```

```python
# load the trained model,scaler pickle,ohe
```

```python
model=load_model('model.h5')
```

```python
# load the encoder and scaler
```

```python
with open('le.pkl','rb') as file:
    labelencoder_gender=pickle.load(file)

with open('ohe_geo.pkl','rb') as file:
    ohe_geography=pickle.load(file)
    
with open('sscaler.pkl','rb') as file:
    standard_scaler=pickle.load(file)
```

```python
input_data = {
    'CreditScore' : 600,
    'Geography' : 'France',
    'Gender' : 'Male',
    'Age' : 40,
    'Tenure' : 3,
    'Balance' : 60000,
    'NumOfProducts' : 2,
    'HasCrCard' : 1,
    'IsActiveMember' : 1,
    'EstimatedSalary' : 50000
} 
```

```python
geo_encoded=ohe_geography.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=ohe_geography.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
# converting dict to df 
input_data_df=pd.DataFrame([input_data])
input_data_df
```

```python
# combining ohe columns with input data
input_data=pd.concat([input_data_df.reset_index(drop=True),geo_encoded_df],axis=1)
input_data
```

```python
input_data['Gender']=labelencoder_gender.transform(input_data['Gender'])
```

```python
input_df=input_data.drop(['Geography'],axis=1)
input_df
```

```python
# scaling the input data
input_scaled=standard_scaler.transform(input_df)
input_scaled
```

```python
prediction=model.predict(input_scaled)
prediction
```

```python
probability=prediction[0][0]
probability
```

```python
if probability > 0.5:
    print("The customer is likely to churn")
else:
    print("The customer is not likely to churn")
```
